```python
# Open image files
#path = "/content/drive/MyDrive/HIDS509 - Imaging Informatics/HIDS509 - Final Project/data/Rembrandt_T1/"
#dirs = os.listdir( path )

# This would print all the files and directories
#for imageFiles in dirs:
   #print(imageFiles)

def getListOfFiles(imagePath):
    # create a list of file and sub directories 
    # names in the given directory 
    listOfFile = os.listdir(imagePath)
    allFiles = list()
    # Iterate over all the entries
    for entry in listOfFile:
        # Create full path
        fullPath = os.path.join(imagePath, entry)
        # If entry is a directory then get the list of files in this directory 
        if os.path.isdir(fullPath):
            allFiles = allFiles + getListOfFiles(fullPath)
        else:
            allFiles.append(fullPath)
                
    return allFiles
```
